```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('dataset/train_emoji.csv', header=None)
test = pd.read_csv('dataset/test_emoji.csv', header=None)
```

```python
X_train = train[0]
Y_train = train[1]

X_test = test[0]
Y_test = test[1]
```

```python
for i in range(5):
    print(X_train[i], emoji.emojize(emoji_dictionary[str(Y_train[i])]))
```

```python
f = open('glove.6B.50d.txt', encoding='utf-8')
```

```python
embeddings_index = {}
for line in f:
    values = line.split() 
    word = values[0]
    coefs = np.asarray(values[1:],dtype='float')
    embeddings_index[word] = coefs

f.close()
```

```python
emb_dim = embeddings_index["play"].shape[0]
print(emb_dim)
```

```python
def embedding_output(X):
    maxLen = 10
    # max no. of words in a sentence is 10
    embedding_out = np.zeros((X.shape[0],maxLen,emb_dim))
    
    for i in range(X.shape[0]):
        X[i] = X[i].split()
        
        for j in range(len(X[i])):
            try:
                embedding_out[i][j] = embeddings_index[X[i][j].lower()]
            except:
                embedding_out[i][j] = np.zeros((50,))
    
    return embedding_out
    
```

```python
embedding_matrix_train = embedding_output(X_train)
embedding_matrix_test = embedding_output(X_test)
```

```python
print(X_train[0])
print(len(X_train[0]))
```

```python
print(embedding_matrix_train.shape)
print(embedding_matrix_test.shape)
```

```python
Y_train = to_categorical(Y_train, num_classes=5)
Y_test = to_categorical(Y_test, num_classes=5)

print(Y_train.shape)
print(Y_train[0])
```

```python
from keras.models import Sequential
from keras.layers import *
```

```python
model = Sequential()
model.add(LSTM(64, input_shape=(10,50), return_sequences=True))
# 64 is the size of hidden state
model.add(Dropout(0.5))
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(5))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
from keras.utils import to_categorical
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint("best_model.h5", monitor='val_loss', mode='min', verbose=True, save_best_only=True)
earlystop = EarlyStopping(monitor='val_acc', patience=10)

hist = model.fit(embedding_matrix_train, Y_train, epochs=150, batch_size=64, shuffle=True, validation_split=0.2)
```

```python
pred_ = model.predict(embedding_matrix_test)
pred = np.argmax(pred_, axis=1)
```
